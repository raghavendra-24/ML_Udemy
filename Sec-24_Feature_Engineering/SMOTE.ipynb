## SMOTE (Synthetic Minority Oversampling Technique)

SMOTE (Synthetic Minority Oversampling Technique) is a data augmentation technique used in machine learning to handle imbalanced datasets. It works by creating synthetic samples of the minority class by interpolating between existing samples. This helps improve model performance by providing a more balanced dataset for training, reducing bias towards the majority class.

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_redundant=0,n_features=2,n_clusters_per_class=1,
                        weights=[0.90],random_state=12)
```

```python
import pandas as pd
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2],axis=1)
final_df.head()
```

```python

final_df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'],final_df['f2'],c=final_df['target'])
```

```python
# transfrom the dataset
oversample=SMOTE()
X,y= oversample.fit_resample(final_df[['f1','f2']],final_df['target'])
```
